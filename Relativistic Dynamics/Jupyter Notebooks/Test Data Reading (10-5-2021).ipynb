```python
import numpy as np
import pandas as pd
import os

import matplotlib as mpl
import matplotlib.pyplot as plt
import modplot
```

```python
os.listdir()
```

```python
data_dir = 'test3.Spe'
```

```python
f = open(data_dir, 'r')
data = f.readlines()
```

```python
counts = []
for line in data[12:2060]:
    newline = line.replace('\n', ' ')
    count = float(newline)
    counts.append(count)
```

```python
counts
```

```python
max(counts)
```

```python
plt.plot(counts)
plt.xlim(0,1000)
plt.yscale('log')
```

```python
mpl.rcParams['figure.dpi'] = 300
mpl.rc('text', usetex = True)
mpl.rc('font', family = 'serif')
```

Linear Scale

```python
fig, [ax] = modplot.axes(ratio_plot=False, xlabel='Bins', ylabel='Counts', xlim=(0,500), 
                      ylim=(0,20000))
ax.plot(counts, color='red', label='Run 1')
ax.legend(loc='upper right')
modplot.stamp(0.55, 0.5, ax=ax, delta_y=0.05,
             line_0 = '$^{133}Ba$ Spectrum')
```

Logarithmic Scale

```python
fig, [ax] = modplot.axes(ratio_plot=False, xlabel='Bins', ylabel='Counts', xlim=(0,1000), 
                      ylim=(10**-1,10**5))
ax.set_yscale('log')
ax.plot(counts, color='red', label='Run 1', lw=0.5)
ax.plot()
ax.legend(loc='upper right')
modplot.stamp(0.55, 0.5, ax=ax, delta_y=0.05,
             line_0 = '$^{133}Ba$ Spectrum')
```
