**Make Plots of the Stopping Voltages**

Get the Stopping Voltage and Frequency Data

```python
# frequencies (GHz)
freqs = [821.92, 741.29, 688.39, 549.35, 519.12]

# stopping voltgaes (V)
stopvolts = [1.18, 0.77, 0.68, 0.29, 0.18]

# errors
errs_sv = [0.1603, 0.1603, 0.1603, 0.1603, 0.1603]
```

```python
np.std(stopvolts)/(5**0.5)
```

```python
def chisq(x,y):
    '''
    Inputs:
    x = x data
    y = y data
    sig = sigma data
    Outputs:
    chi2 = chi-squared value
    '''
    expected_vals = 0.003132*np.array(x)-1.46
    return np.sum((np.array(y)-expected_vals)**2/expected_vals)

chi2 = chisq(freqs, stopvolts)
```

```python
chi2
```

```python
fig, [ax] = modplot.axes(ratio_plot=False, xlabel='Frequency [GHz]',
                         ylabel='Stopping Voltage [V]',
                         xlim=(500, 850),
                         ylim=(0, 1.5))

# data
ax.errorbar(freqs, stopvolts, xerr=0, yerr=errs_sv, marker='.', color='blue', 
            label='Data', ms=1, ls='none', lw=0.5, elinewidth=0.5, capthick=1, capsize=1)

# linear fit
x = np.linspace(500,850,200)
y = 31.32*x - 1.46
ax.plot(x, 0.003132*x-1.46, color = 'red', lw = 0.5, label='Linear Fit')

ax.legend(loc='upper right', frameon=False, numpoints=2)

modplot.stamp(0.05, 0.55, ax=ax, delts_y=0.055, textops_update={'fontsize':12}, 
              line_0=r'$V_{\mathrm{stop}}(\nu)=\frac{h}{e}\nu-\frac{\phi}{e}$',
              line_1='$R^{2}=0.9655$')

modplot.stamp(0.20, 0.15, ax=ax, delta_y=0.065,textops_update = {'fontsize':12},
              line_0='$h = (5.0112 \pm 1.6905)\cdot 10^{-34}$ Js',
              line_1='$\phi = (-2.3406 \pm 0.0160)\cdot 10^{-19}$ J')

ax.set_title(label='Stopping Voltages for Filter Frequency', loc='center')

save = True
if save:
    plt.savefig(dir_plots_save + '/stopping_volts.png')
```
