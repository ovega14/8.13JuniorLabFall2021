Load in the Data

```python
os.chdir(dir_data)
```

```python
f = open('11-9-2021_TsungDaoLee_run1.txt', 'r')
```

```python
lines = f.readlines()
```

```python
lines[7].split()[1][5:7]
```

```python
# get the times and incidence counts

times = []
counts = []

for line in lines[7:]:
    text = line.split()
    
    time = text[1]
    count = float(text[10])
    
    times.append(time)
    counts.append(count)
```

```python
times
```

```python
# reformat the times
stamps = []

for time in times:
    minute = float(time[2:4])-float(times[0][2:4])
    second = float(time[5:7])-float(times[0][5:7])
    hour = float(time[0:1])-float(times[0][0:1])
    
    newtime = 3600*hour + 60*minute + second
    
    stamps.append(newtime)
```

**Plot the Data**

```python
# plotting parameters
save = False
mpl.rcParams['figure.dpi'] = 300
mpl.rc('text', usetex = True)
mpl.rc('font', family = 'serif')
```

```python
plt.plot(stamps, np.cumsum(counts), color='orange')
plt.ylabel('Accrued Counts')
plt.xlabel('Time [s]')
plt.savefig(dir_plot + 'accrued1')
```

```python
# get count rates
len(stamps)
```

```python
plt.plot(stamps, np.array(np.cumsum(counts))/np.array(stamps), color='purple')
plt.ylabel('Average Count Rate [counts/s]')
plt.xlabel('Time [s]')
plt.savefig(dir_plot + 'rate1')
```
